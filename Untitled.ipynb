```python
import numpy as np
from sympy.core.symbol import symbols
from sympy.solvers.solveset import *
from scipy import integrate

from matplotlib import pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from matplotlib.colors import cnames
from matplotlib import animation
```

```python
A1,A2,B,C=1,1,5,5

def toggle_deriv(X, t=0):
    """Compute the time-derivative of a toggle switch system."""
    x,y = X[0],X[1]
    return [A1 / (1 + y**B) - x, A2 / (1 + x**C) - y]

t = np.linspace(0, 50,  10000)              # time
X0 = np.random.rand(2)                # initials conditions: 10 rabbits and 5 foxes
X, infodict = integrate.odeint(toggle_deriv, X0, t, full_output=True)
infodict['message']                     # >>> 'Integration successful.'
```

```python
x, y = X.T
fig = plt.figure()
plt.plot(t, x, 'r-', label='x')
plt.plot(t, y  , 'b-', label='y')
plt.grid()
plt.legend(loc='best')
plt.xlabel('time')
plt.ylabel('concentration')
plt.title('Evolution of x,y concentrations')
```

```python
fig2 = plt.figure()
plt.plot(X[:,0],X[:,1])
plt.xlim(0,1)
plt.ylim(0,1)
```

```python
values  = np.linspace(0.1, 0.9, 5)                          # position of X0 between X_f0 and X_f1
vcolors = plt.cm.autumn_r(np.linspace(0.3, 1., len(values)))  # colors for each trajectory

fig2 = plt.figure()

#-------------------------------------------------------
# plot trajectories
for v, col in zip(values, vcolors):
    X0 = np.random.rand(2)
    X = integrate.odeint( toggle_deriv, X0, t)         # we don't need infodict here
    plt.plot( X[:,0], X[:,1], lw=3.5*v, color=col, label='X0=(%.f, %.f)' % ( X0[0], X0[1]) )

#-------------------------------------------------------
# define a grid and compute direction at each point
# ymax = plt.ylim(ymin=0)[1]                        # get axis limits
# xmax = plt.xlim(xmin=0)[1]
nb_points   = 20

x = np.linspace(0, 1, nb_points)
y = np.linspace(0, 1, nb_points)

X1 , Y1  = np.meshgrid(x, y)                       # create a grid
DX1, DY1 = toggle_deriv([X1, Y1])                      # compute growth rate on the gridt
M = (np.hypot(DX1, DY1))                           # Norm of the growth rate 
M[ M == 0] = 1.                                 # Avoid zero division errors 
DX1 /= M                                        # Normalize each arrows
DY1 /= M

#-------------------------------------------------------
# Drow direction fields, using matplotlib 's quiver function
# I choose to plot normalized arrows and to use colors to give information on
# the growth speed
plt.title('Trajectories and direction fields')
Q = plt.quiver(X1, Y1, DX1, DY1, M, pivot='mid', cmap=plt.cm.jet)
plt.xlabel('Conc. of x')
plt.ylabel('Conc. of x')
plt.legend()
plt.grid()
plt.xlim(0, 1)
plt.ylim(0, 1)
#f2.savefig('rabbits_and_foxes_2.png')
```

```python
def solve_toggle(N=10, max_time=5.0, A1=1,A2=1,B=5,C=5,VecField=False):
    fig=plt.figure()
    ax = fig.add_axes([0,0,1,1])
    #ax.axis('off')
    # prepare the axes limits
    ax.set_xlim((0, 1))
    ax.set_ylim((0, 1))
    def toggle_deriv(X, t=0):
        """Compute the time-derivative of a toggle switch system."""
        x,y = X[0],X[1]
        return [A1 / (1 + y**B) - x, A2 / (1 + x**C) - y]
    if VecField:
        nx = np.linspace(0,1,20)
        X1 , Y1  = np.meshgrid(nx, nx)      # create a grid
        DX1, DY1 = toggle_deriv([X1, Y1])    # compute growth rate on the grid
        M = (np.hypot(DX1, DY1))             # Norm of the growth rate 
        M[ M == 0] = 1.                     # Avoid zero division errors 
        DX1 /= M                          # Normalize each arrows
        DY1 /= M
        plt.quiver(X1, Y1, DX1, DY1, M, pivot='mid', cmap=plt.cm.viridis)
    
    # Choose random starting points, uniformly distributed from -15 to 15
    #np.random.seed(1)
    X0 = np.random.rand(N, 2)

    # Solve for the trajectories
    t = np.linspace(0, max_time, int(250*max_time))
    X_t = np.asarray([integrate.odeint(toggle_deriv, X0_i, t)
                      for X0_i in X0])

    # choose a different color for each trajectory
    colors = plt.cm.viridis(np.linspace(0, 1, N))

    for i in range(N):
        x, y = X_t[i,:].T
        lines = ax.plot(x, y, '-', c=colors[i])
        plt.setp(lines, linewidth=2)

    plt.xlabel('conc. of x')
    plt.ylabel('conc. of y')
    plt.grid()
    #ax.view_init(30, angle)
    plt.show()

    return t, X_t
```

```python
def solve_lorenz(N=10, angle=0.0, max_time=4.0, sigma=10.0, beta=8./3, rho=28.0):

    fig = plt.figure()
    ax = fig.add_axes([0, 0, 1, 1], projection='3d')
    ax.axis('off')

    # prepare the axes limits
    ax.set_xlim((-25, 25))
    ax.set_ylim((-35, 35))
    ax.set_zlim((5, 55))

    def lorenz_deriv(x_y_z, t0, sigma=sigma, beta=beta, rho=rho):
        """Compute the time-derivative of a Lorenz system."""
        x, y, z = x_y_z
        return [sigma * (y - x), x * (rho - z) - y, x * y - beta * z]

    # Choose random starting points, uniformly distributed from -15 to 15
    np.random.seed(1)
    x0 = -15 + 30 * np.random.random((N, 3))

    # Solve for the trajectories
    t = np.linspace(0, max_time, int(250*max_time))
    x_t = np.asarray([integrate.odeint(lorenz_deriv, x0i, t)
                      for x0i in x0])

    # choose a different color for each trajectory
    colors = plt.cm.viridis(np.linspace(0, 1, N))

    for i in range(N):
        x, y, z = x_t[i,:,:].T
        lines = ax.plot(x, y, z, '-', c=colors[i])
        plt.setp(lines, linewidth=2)

    ax.view_init(30, angle)
    plt.show()

    return t, x_t
```

```python
t, x_t = solve_lorenz(angle=0, N=10)
```
