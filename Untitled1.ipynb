```python
from Hysteresis import *
import random
from matplotlib.animation import FuncAnimation
from IPython.display import HTML
%config InlineBackend.figure_formats = {'png', 'retina'}
```

```python
init_cond = [.5,0]
t_final = 30
eps = .1
rhs = lambda t, y: drift_rhs(t,y,eps)
```

```python
plot_solution(rhs,init_cond,t_final)
```

```python
plot_trajectory(None,rhs,init_cond,t_final)
```

```python
streamplot(None,rhs,[-2,2],[-2,2])
```

```python
plot_nullclines(None,[-2,2],[-2,2])
```

```python
# Set up the figure
fig, ax = plt.subplots()
ax.set_xlabel('c')
ax.set_ylabel('y')
# Build the plot
c_bounds = [-0.5,1.5]
y_bounds = [-2,2]
#ax.set_aspect('equal')
ax.axis([c_bounds[0],c_bounds[1],y_bounds[0],y_bounds[1]])
streamplot(ax,rhs,c_bounds,y_bounds)
plot_nullclines(ax,c_bounds,y_bounds)
plot_fixedpoints(ax,eps,c_bounds,y_bounds)
N=20
for i in range(N):
    rand_c = random.uniform(c_bounds[0],c_bounds[1])
    rand_y = random.uniform(y_bounds[0],y_bounds[1])
    init = np.array([rand_c,rand_y])
    plot_trajectory(ax,rhs,init,t_final,1)
```

```python
fig,ax = plt.subplots()
ax.grid()
c_bounds = [-0.5,1.5]
y_bounds = [-2,2]
t_max = 100
t_bounds = [0,t_max]
ax.set(xlim=(c_bounds[0],c_bounds[1]),ylim=(y_bounds[0],y_bounds[1]))
plot_nullclines(ax,c_bounds,y_bounds)
n=20
m = 100

c = np.linspace(c_bounds[0],c_bounds[1], n)
y = np.linspace(y_bounds[0],y_bounds[1], n)
t = np.linspace(t_bounds[0],t_bounds[1], 100)
```

```python
cv,yv,tv = np.meshgrid(c,y,t)
U = .0005*tv*(tv-t_max)
V = cv+yv-yv**3
```

```python
i = 30
q = ax.quiver(cv[...,0],yv[...,0],U[...,0],V[...,0])
fig
```

```python
def animate(i):
    q.set_UVC(U[..., i], V[..., i])
```

```python
anim = FuncAnimation(fig, animate, interval=100, frames=len(t)-1)
```
